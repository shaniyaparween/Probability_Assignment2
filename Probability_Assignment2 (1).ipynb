```python
import random

# Simulate 10,000 coin tosses
tosses = [random.choice(['Heads', 'Tails']) for _ in range(10000)]
heads = tosses.count('Heads')
tails = tosses.count('Tails')

print(f"Experimental Probability of Heads: {heads / 10000}")
print(f"Experimental Probability of Tails: {tails / 10000}")
```

```python
sum_7_count = 0
for _ in range(10000):
    d1, d2 = random.randint(1, 6), random.randint(1, 6)
    if d1 + d2 == 7:
        sum_7_count += 1

print(f"Probability of getting sum = 7: {sum_7_count / 10000}")

```

```python
def estimate_prob_at_least_one_6(trials=10000):
    count = 0
    for _ in range(trials):
        if 6 in [random.randint(1, 6) for _ in range(10)]:
            count += 1
    return count / trials

prob = estimate_prob_at_least_one_6()
print(f"Estimated Probability of at least one 6 in 10 rolls: {prob}")
```

```python
colors = ['Red'] * 5 + ['Green'] * 7 + ['Blue'] * 8

# Simulate 1000 draws with replacement
draws = [random.choice(colors) for _ in range(1000)]

# Count pairs where previous was Blue and current is Red
blue_red_pairs = 0
blue_prev_count = 0

for i in range(1, len(draws)):
    if draws[i-1] == 'Blue':
        blue_prev_count += 1
        if draws[i] == 'Red':
            blue_red_pairs += 1

if blue_prev_count > 0:
    cond_prob = blue_red_pairs / blue_prev_count
    print(f"P(Red | Previous was Blue) ≈ {cond_prob}")
else:
    print("No Blue found as previous. Cannot compute.")
```

```python
# Marginal probabilities
P_R = 5 / 20
P_B = 8 / 20
P_B_given_R = sum([1 for i in range(1, len(draws)) if draws[i-1]=='Red' and draws[i]=='Blue']) / sum([1 for i in range(1, len(draws)) if draws[i-1]=='Red'])

# Bayes Theorem: P(R|B) = (P(B|R) * P(R)) / P(B)
P_R_given_B_est = (P_B_given_R * P_R) / P_B
print(f"P(R|B) using Bayes' Theorem ≈ {P_R_given_B_est}")
```

```python
import numpy as np

values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.4]
sample = np.random.choice(values, size=1000, p=probabilities)

print(f"Empirical Mean: {np.mean(sample)}")
print(f"Empirical Variance: {np.var(sample)}")
print(f"Empirical Std Dev: {np.std(sample)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = np.random.exponential(scale=5, size=2000)

# Histogram with KDE
plt.figure(figsize=(8,5))
sns.histplot(data, bins=30, kde=True, stat='density', color='skyblue')
plt.title('Exponential Distribution (λ=1/5)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
# Generate 10,000 random numbers from uniform distribution
uniform_data = np.random.uniform(0, 1, 10000)

# Draw 1000 samples of size 30, and compute their means
sample_means = [np.mean(np.random.choice(uniform_data, size=30)) for _ in range(1000)]

# Plotting
plt.figure(figsize=(10, 4))

# Original Uniform Distribution
plt.subplot(1, 2, 1)
sns.histplot(uniform_data, bins=30, kde=True, color='orange')
plt.title("Original Uniform Distribution")

# Distribution of Sample Means
plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=30, kde=True, color='green')
plt.title("Distribution of Sample Means (CLT)")

plt.tight_layout()
plt.show()
```
